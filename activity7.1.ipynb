```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import folium
from folium.plugins import HeatMap
```

```python
df = pd.read_csv('datasets\\Global_terrorist.csv', encoding='ISO-8859-1')
```

```python
df
```

<h1> Filtering Data</h1>

```python
filter1 = df[df['region_txt'] == "Middle East & North Africa"]
```

```python
filter2 = filter1[filter1['attacktype1_txt'] == 'Bombing/Explosion']
```

```python
filter3 = filter2[filter2['nkill'] >= 10]
```

```python
filter4 = filter3[filter3['iyear'] >= 2000]
```

```python
filter4
```

```python
ME_terror = filter4.groupby(['longitude', 'latitude']).size().reset_index(name='incident_count')

ME_terror
```

<h1>VISUALIZATION OF GEOSPATIAL USING FOLIUM</h1>

```python
heat_globl= ME_terror[[ 'latitude', 'longitude','incident_count']].values.tolist()
```

```python
my_map = folium.Map(location = [30,40], zoom_start = 4)
HeatMap(heat_globl).add_to(my_map)
my_map
```

<h1>CRIME/ATTACK ANALYSIS</h1>

```python
# Filter for region
region_data = df[df['region_txt'] == 'Western Europe']

# Group by location and count attacks
crime_grouped = region_data.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')
crime_grouped.columns = ['latitude', 'longitude', 'incident_count']
crime_grouped = crime_grouped.dropna()

# HeatMap visualization

heat_globl = crime_grouped[['latitude', 'longitude', 'incident_count']].values.tolist()
crime_map = folium.Map(location=[30, 40], zoom_start=4)
HeatMap(heat_globl).add_to(crime_map)

crime_map
```

<h1> CASUALITY ANALYSIS</h1>

```python
# Filter for region and valid kill data
casualty_data = df[(df['region_txt'] == 'Southeast Asia') & (df['nkill'].notnull())]

# Group by location and sum killed
casualty_grouped = casualty_data.groupby(['latitude', 'longitude'], as_index=False)['nkill'].sum()
casualty_grouped = casualty_grouped.dropna()
casualty_grouped.rename(columns={'nkill': 'incident_count'}, inplace=True)

# HeatMap visualization
heat_globl = casualty_grouped[['latitude', 'longitude', 'incident_count']].values.tolist()
casualty_map = folium.Map(location=[30, 40], zoom_start=4)
HeatMap(heat_globl).add_to(casualty_map)

casualty_map
```

<h1>NATIONALITY ANALYSIS </h1>

```python
# Filter for region and valid nationality data
nationality_data = df[(df['region_txt'] == 'Sub-Saharan Africa') & (df['natlty1_txt'].notnull())]

# Group by location and count nationality-based incidents
nationality_grouped = nationality_data.groupby(['latitude', 'longitude'], as_index=False).size()
nationality_grouped.columns = ['latitude', 'longitude', 'incident_count']
nationality_grouped = nationality_grouped.dropna()

# HeatMap visualization
heat_globl = nationality_grouped[['latitude', 'longitude', 'incident_count']].values.tolist()
nationality_map = folium.Map(location=[30, 40], zoom_start=4)
HeatMap(heat_globl).add_to(nationality_map)

nationality_map
```
